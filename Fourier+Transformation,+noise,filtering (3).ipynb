```python
import scipy as sp
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Fs = 1000
t = np.arange(0,2,1./Fs)
x1= np.cos(2*np.pi*2*t) +2*np.cos(2*np.pi*5*t)
n= np.random.randn(len(x1))*0.5
x = x1+n
plt.plot(t,x)
plt.title('Signal with noise')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')


```

```python
X = sp.fft(x)
F = sp.linspace(0,Fs,len(x)+1)
F = F[:-1]
N = len(X)
plt.figure()
plt.stem(F[0:100], sp.absolute(X[0:100]), 'g')
plt.title('Signal spectrum')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Amplitude')
```
