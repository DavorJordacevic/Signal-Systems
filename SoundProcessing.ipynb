```python
#In this excercise we will process the sound signal. Upload some wav file from https://freesound.org/


import scipy as sp
from scipy.io.wavfile import read
from scipy.io.wavfile import write
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
(fs,x)=read('monosignal.wav')
```

```python
fs
```

```python
len(x)
```

```python
plt.plot(x)
```

```python
X=sp.fft(x)
```

```python
f=np.linspace(0,fs,len(x))
```

```python
plt.stem(f[0:500],np.abs(X[0:500]))
```

```python
from scipy import signal
b,a=signal.butter(5,100/(fs/2),btype='lowpass')
```

```python
filteredsignal=signal.lfilter(b,a,x)
```

```python
write('filteredsignal.wav',fs,filteredsignal) #possible to open with aimp3 software
```

```python
noise=np.random.randn(len(x))
```

```python
signalwithnoise=x+noise
```

```python
plt.plot(signalwithnoise)
```

```python
write('signalwithnoise.wav',fs,signalwithnoise)
```
