```python
import matplotlib.pyplot as plt 
import numpy as np
import scipy as sp
from scipy import signal 
% matplotlib inline
```

```python
t = np.arange(0,2,0.001)
x = np.sin(2*np.pi*2*t) + 2*np.sin(2*np.pi*7*t) + 3*np.sin(2*np.pi*11*t) + 0.5*np.random.randn(len(t))
plt.plot(t,x)
```

```python
Fs=1000
```

```python
# Filtriranje signala Batervortovim filtrom
```

```python
# Filter propusnik niskih ucestanosti - lowpass
b,a = signal.butter(4,0.026,btype='lowpass')
w,h = signal.freqz(b,a)
fNorm = w / sp.pi
fNew = fNorm * (Fs/2)
plt.plot(fNew[0:50],sp.absolute(h[0:50]))
```

```python
y = signal.lfilter(b,a,x)
plt.plot(t,y)
```

```python
# Furijeova transformacija
```

```python
Y = sp.fft(y)
f = np.linspace(0,Fs,len(y))
plt.stem(f[0:50],np.abs(Y[0:50]))
```

```python
# Filter propusnik visokih ucestanosti - lowpass
b,a = signal.butter(4,0.02,btype='highpass')
w,h = signal.freqz(b,a)
fNorm = w / sp.pi
fNew = fNorm * (Fs/2)
plt.plot(fNew[0:50],sp.absolute(h[0:50]))
```

```python
z = signal.lfilter(b,a,x)
plt.plot(t,z)
```

```python
# Furijeova transformacija
Z = sp.fft(z)
f = np.linspace(0,Fs,len(z))
plt.stem(f[0:50],np.abs(Z[0:50]))
```
