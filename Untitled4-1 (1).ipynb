```python
# Filtriranje signala
# Batervortov filter
```

```python
b, a = signal.butter(5,0.01 , btype ='low')
w, h = signal.freqz(b,a)
fNorm = w / sp.pi
f1 = fNorm * (Fs/2)
```

```python
plt.plot(f1, sp.absolute(h))
plt.title('Filter')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
```

```python
y = signal.lfilter(b,a,x)
plt.plot(t,y)
```

```python
Y=sp.fft(y)
Fs=1000
f=np.linspace(0,Fs,len(y)+1)
f=f[:-1]
n=len(y)
plt.stem(f[0:50],np.abs(Y[0:50]))
```
