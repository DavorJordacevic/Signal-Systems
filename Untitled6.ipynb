```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy import signal
%matplotlib inline
```

```python
t = np.arange(0,2,0.001)
x = np.sin(2*np.pi*2*t) +3*np.sin(2*np.pi*7*t) + 2*np.sin(2*np.pi*11*t) + 0.5*np.random.randn(len(t))
plt.plot(t,x)
```

```python
n=0.25*np.random.randn(len(t))
plt.plot(t,n)
```

```python
plt.hist(n,bins=200)
```

```python
# furijeova transformacija
```

```python
X = sp.fft(x)
```

```python
fs = 1000
```

```python
f = np.linspace(0,fs,len(x))
```

```python
f
```

```python
plt.stem(f[0:50],np.abs(X[0:50]))
```

```python
# Filtriranje signala
```

```python
b ,a = signal.butter(4,0.024 , btype='lowpass')
```

```python
w , h = signal.freqz(b,a)
```

```python
fNorm = w / sp.pi
f1 = fNorm * (fs/2)
```

```python
plt.plot(f1[0:50], sp.absolute(h[0:50]))
```

```python
y=signal.lfilter(b,a,x)
```

```python
plt.plot(t,y)
```

```python
Y= sp.fft(y)
f = np.linspace(0,fs,len(y))
```

```python
plt.stem(f[0:50],np.abs(Y[0:50]))
```
