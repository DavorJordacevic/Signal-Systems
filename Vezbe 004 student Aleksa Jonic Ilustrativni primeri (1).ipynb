```python
import scipy as sp
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
```

```python
Fs = 100
t = sp.arange(0,2,1/Fs)
x = sp.cos(2*sp.pi*t*2) + sp.cos(2*sp.pi*t*11) + sp.cos(2*sp.pi*t*22)
plt.plot(t,x,'r')
```

```python
X = sp.fft(x)
F = sp.linspace(0,Fs,len(x)+1)
F = F[:-1]
N = len(X)
plt.stem(F[0:N/2], sp.absolute(X[0:N/2]), 'r')
```

```python
#low pass filter koji ce propustiti samo signal na 2Hz
b,a = signal.butter(5,0.1, btype = 'low')
w,h = signal.freqz(b,a)
fNorm = w / sp.pi
fNew = fNorm * (Fs/2)
```

```python
plt.plot(fNew, sp.absolute(h), 'g')
```

```python
y = signal.lfilter(b,a,x)
plt.subplot(211)
plt.plot(t,x, 'b')
plt.subplot(212)
plt.plot(t,y, 'r')
```

```python
Y = sp.fft(y)
F = sp.linspace(0,Fs, len(Y)+1)
F = F[:-1]
N = len(Y)
plt.stem(F[0:N/2], sp.absolute(Y[0:N/2]),'g')
```

```python
#Hocemo da propustimo samo signal na 11Hz
b,a = signal.butter(5,[0.1, 0.3], btype = 'bandpass')
w,h = signal.freqz(b,a)
fNorm = w / sp.pi
fNew = fNorm * (Fs/2)
plt.plot(fNew, sp.absolute(h), 'g')
```

```python
y2 = signal.lfilter(b,a,x)
plt.plot(t,y2,'r')
```

```python
Y2 = sp.fft(y2)
F = sp.linspace(0,Fs, len(Y2)+1)
F = F[:-1]
N = len(Y2)
plt.stem(F[0:N/2], sp.absolute(Y2[0:N/2]),'b')
```

```python
plt.subplot(211)
plt.plot(t,x, 'b')
plt.subplot(212)
plt.plot(t,y2, 'r')
```

```python
#hocemo da isfiltriramo samo signal na 11Hz
b,a = signal.butter(5,[0.125,0.312], btype = 'bandstop')
w,h = signal.freqz(b,a)
fNorm = w / sp.pi
fNew = fNorm * (Fs/2)
plt.plot(fNew, sp.absolute(h))
```

```python
y3 = signal.lfilter(b,a,x)
plt.plot(t,y3)
```

```python
Y3 = sp.fft(y3)
F = sp.linspace(0,Fs, len(Y3)+1)
F = F[:-1]
N = len(Y3)
plt.stem(F[0:N/2], sp.absolute(Y3[0:N/2]),'b')
```
